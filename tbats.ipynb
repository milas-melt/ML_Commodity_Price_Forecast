```python
from tbats import TBATS, BATS
import tool
```

```python
df,_ = tool.data_process("IndexPrices__US2792.xlsx")
df.drop(columns=['cap', 'floor'], inplace=True)
df.columns = ['ds', 'value']
```

```python
estimator = TBATS(seasonal_periods=(7.15, 365.25))
model = estimator.fit(df.value)
y_forecast = model.forecast(steps=548)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(_.y, y_forecast)
```
